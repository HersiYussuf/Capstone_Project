```python
# loading the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Histogram for 'time_in_hospital'
plt.figure(figsize=(8, 6))
df['time_in_hospital'].hist(bins=20)
plt.title('Time in Hospital Distribution')
plt.xlabel('Time in Hospital')
plt.ylabel('Count')
plt.show()
```

```python
# Pie chart for 'diabetesMed'
plt.figure(figsize=(6, 6))
df['diabetesMed'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Diabetes Medication')
plt.ylabel('')
plt.show()
```

```python
# Pie chart for 'readmitted'
plt.figure(figsize=(6, 6))
df['readmitted'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Readmission Status')
plt.ylabel('')
plt.show()
```

```python
# Box plot for 'num_lab_procedures' across 'race'
plt.figure(figsize=(8, 6))
df.boxplot(column='num_lab_procedures', by='race')
plt.title('Number of Lab Procedures by Race')
plt.xlabel('Race')
plt.ylabel('Number of Lab Procedures')
plt.show()
```

```python


df.groupby(['race', 'gender']).size().unstack().plot(kind='bar')
plt.xlabel('Race')
plt.ylabel('Count')
plt.title('Count of Patients by Race and Gender')
plt.show()
```

**Stacked Bar Chart: If you have multiple categorical variables that you want to compare together, you can create a stacked bar chart. For example, you can stack the bar chart for 'race' and 'gender' to see the distribution of gender within each race category.**

```python
# Stacked bar chart for 'race' and 'gender'
race_gender_counts = df.groupby(['race', 'gender']).size().unstack()
race_gender_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribution of Gender within Race Categories')
plt.xlabel('Race')
plt.ylabel('Count')
plt.show()

# Repeat the same process for other combinations of categorical variables
```

**Violin Plot: If you want to visualize the distribution of a numerical variable across different categories, you can create a violin plot. This plot combines a box plot and a kernel density plot to show the distribution and the density estimation.**

```python
# Violin plot for 'time_in_hospital' across 'readmitted'
plt.figure(figsize=(8, 6))
sns.violinplot(x='readmitted', y='time_in_hospital', data=df)
plt.title('Time in Hospital Distribution for Readmission Status')
plt.xlabel('Readmission Status')
plt.ylabel('Time in Hospital')
plt.show()

# Repeat the same process for other numerical variables and categorical variables of interest
```

```python
# Line chart for 'num_lab_procedures', 'num_procedures', and 'num_medications'
plt.figure(figsize=(10, 6))
plt.plot(df['num_lab_procedures'], label='Number of Lab Procedures')
plt.plot(df['num_procedures'], label='Number of Procedures')
plt.plot(df['num_medications'], label='Number of Medications')
plt.title('Trends of Medical Procedures and Medications')
plt.xlabel('Sequence')
plt.ylabel('Count')
plt.legend()
plt.show()

# Repeat the same process for other numerical variables of interest
```

Pair Plot (Scatter Matrix): If we have multiple numerical variables and want to explore the relationships between them,we can use a pair plot, also known as a scatter matrix. This plot will show scatter plots for all combinations of numerical variables.

Box Plot with Hue: If we have a numerical variable and want to compare its distribution across different categories, we  can use a box plot with the 'hue' parameter to separate the box plots by a categorical variable.

```python
# Box plot for 'num_medications' with 'gender' as hue
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='num_medications', data=df)
plt.title('Number of Medications by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Medications')
plt.show()

# Repeat the same process for other numerical variables and categorical variables of interest
```

Count Plot: If we want to visualize the count of occurrences for a single categorical variable, we can use a count plot.

Pairwise Correlation Heatmap: If we want to explore the pairwise correlation between all numerical variables in the dataset,we can create a heatmap.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the distribution of patients by race
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='race')
plt.title('Distribution of Patients by Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

The bar plot above represents the distribution of patients by race. It provides insights into the racial composition of the dataset, allowing us to understand the diversity of patients included in the analysis.

```python

# Plotting the gender distribution
plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='gender')
plt.title('Gender Distribution of Patients')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

The bar plot above illustrates the distribution of patients by gender. It provides an overview of the gender composition of the dataset, highlighting the representation of male and female patients.

```python


# Plotting the age group distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='age')
plt.title('Age Group Distribution of Patients')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

The bar plot above showcases the distribution of patients across different age groups. It allows us to observe the age distribution of patients in the dataset, providing insights into the age demographics of the study populati

```python

# Plotting the number of diagnoses
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='number_diagnoses', bins=15, kde=True)
plt.title('Distribution of Number of Diagnoses')
plt.xlabel('Number of Diagnoses')
plt.ylabel('Count')
plt.show()
```

The histogram above represents the distribution of the number of diagnoses for patients. It visualizes the frequency of different diagnosis counts, enabling us to understand the range and distribution of diagnoses among the patients.
